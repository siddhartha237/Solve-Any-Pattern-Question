### Problem 13
Print the following pattern.


```python
def print_star_pattern():
    # List of lines in the pattern
    pattern = [
        "    *    ",
        "   * *   ",
        "  *   *  ",
        " *     * ",
        "*********"
    ]
    
    # Print each line in the pattern
    for line in pattern:
        print(line)

print_star_pattern()
```
